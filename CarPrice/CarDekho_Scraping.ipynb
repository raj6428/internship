```python
# Importing Libraries
import selenium
import pandas as pd
import time
from bs4 import BeautifulSoup

# Importing selenium webdriver 
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait as wait
from selenium.webdriver.common.keys import Keys


# Importing required Exceptions which needs to handled
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException, ElementClickInterceptedException

#Importing requests
import requests

# importing regex
import re

car_type=[]
location=[]
brand=[]
model=[]
version=[]
year=[]
fuel_type=[]
transmission_type=[]
distance_driven=[]
price=[]



list_of_cities=['Chennai','Hyderabad','Delhi NCR','Bangalore','Jaipur','Kolkata','Mumbai','Pune','Ahmedabad']
for city in list_of_cities:
    driver=webdriver.Chrome(r"C:\Users\acer\Downloads\chromedriver_win32 (1)\chromedriver.exe")
    driver.maximize_window()
    url=f'https://www.cardekho.com/used-cars+in+{city}'
    driver.get(url)
    print('Loading Cars in city: ',city)
    time.sleep(5)
  
    #Scrolling till end of page
    find_elem =''
    scroll_from = 0
    scroll_limit = 1500
    while not find_elem:
        time.sleep(2)
        driver.execute_script("window.scrollTo(%d, %d);" %(scroll_from, scroll_from+scroll_limit))
        scroll_from += scroll_limit
        try:
            find_elem = driver.find_element_by_xpath("//div[@class='text-center showMoreButton']/span").text
            print('    scrolling down')
        except NoSuchElementException:
            pass
    
    #listings
    listings=driver.find_elements_by_xpath('//div[@class="holder hover"]')
    time.sleep(2)
    print(f'  No. of listings in {city}: {len(listings)} ')
    
    #Distance
    km_driven=driver.find_elements_by_xpath(f'//div[@class="truncate dotlist"]/span[1]')
    for i in km_driven:
        distance_driven.append(i.text)
    print(f'  listed km_driven of {len(distance_driven)} cars')
    
    #Fuel type
    fuel_types=driver.find_elements_by_xpath(f'//div[@class="truncate dotlist"]/span[2]')
    for i in fuel_types:
        fuel_type.append(i.text)
    print(f'  listed fuel type of {len(fuel_type)} cars')

    #Transmission type
    transmission_types=driver.find_elements_by_xpath(f'//div[@class="truncate dotlist"]/span[3]')
    for i in transmission_types:
        transmission_type.append(i.text)
    print(f'  listed transmission type of {len(transmission_type)} cars')
    
    #Prices
    prices=driver.find_elements_by_xpath('//div[@class="gsc_col-xs-5 text-right price"]/span[1]')
    for i in prices:
        price.append(i.text)
    print(f'  listed prices of {len(price)} cars')
    
    
    #year
    years=driver.find_elements_by_xpath('//div[@class="gsc_col-xs-7  carsName"]')
    for i in years:
        year.append(i.text.split()[0])
    print(f'  listed year of {len(year)} cars')
    year
    
    #brand
    brands=driver.find_elements_by_xpath('//div[@class="gsc_col-xs-7  carsName"]')
    for i in brands:
        brand.append(i.text.split()[1])
    print(f'  listed brands of {len(brand)} cars')

    #version
    versions=driver.find_elements_by_xpath('//div[@class="gsc_col-xs-7  carsName"]/div[1]')
    for i in versions:
        version.append(i.text)
    print(f'  listed versions of {len(version)} cars')
    
    #models
    models=driver.find_elements_by_xpath('//div[@class="gsc_col-xs-7  carsName"]/a')
    for i in models:
        temp=i.text.replace('\n','').split()[2:]
        name=' '.join(temp)
        model.append(name)     
    print(f'  listed models of {len(version)} cars')
    
    #city
    for i in range(len(listings)):
        location.append(city)
    print(f'  listed cities of {len(location)} cars')
    
    #body type
    for i in range(len(listings)):
        car_type.append('-')
    print(f'  listed body types of {len(car_type)} cars as -')
          
    #closing driver    
    driver.close()
    print('driver is closed')
    print('\n\n')
          
print('\n\n\nScraping completed')

#Creating Data Frame
cars=pd.DataFrame({})
cars['car_type']=car_type
cars['location']=location
cars['brand']=brand
cars['model']=model
cars['version']=version
cars['year']=year
cars['fuel_type']=fuel_type
cars['transmission_type']=transmission_type
cars['distance_driven']=distance_driven
cars['price']=price

print('Saving into csv')
cars.to_csv('cardekho.csv')
```
